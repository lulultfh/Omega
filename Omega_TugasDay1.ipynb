### **Soal 5: Fungsi**  
  
Buatlah sebuah fungsi bernama `hitung_diskon` yang menerima dua argumen: `harga_awal` dan `persen_diskon`. Fungsi ini harus mengembalikan harga setelah diskon.

````python
### YOUR CODE HERE ###
def hitung_diskon(harga_baju, diskon):
    return harga_baju * (1 - diskon / 100)
### END OF YOUR CODE ###

harga_baju = 250000
diskon = 15
harga_akhir = hitung_diskon(harga_baju, diskon)
print(f"Harga awal Rp{harga_baju} dengan diskon {diskon}% menjadi Rp{int(harga_akhir)}")
````

## ðŸ“Š **Bagian 2: Exploratory Data Analysis (EDA)**

### **Inspect dataset**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = "https://raw.githubusercontent.com/VenkyAdi/EDA---Projects/refs/heads/main/FedEx%20Logistics%20Performance%20Analysis/SCMS_Delivery_History_Dataset.csv"
df = pd.read_csv(data)

# Tampilkan 5 baris pertama dan info singkat
print("Lima baris pertama:", df.head())


print("\nInformasi DataFrame:", df.info())
````

### **Penanganan missing values**

````python
print("Missing values sebelum:")
print(df.isnull().sum())

### YOUR CODE HERE ###
df['Shipment Mode'] = df['Shipment Mode'].fillna('Unknown')
df['Dosage'] = df['Dosage'].fillna('Unknown')
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(0)
### END OF YOUR CODE ###

print("\nMissing values setelah diisi:")
print(df.isnull().sum())
````

### **Analisis Univariat**
Peserta bebas memilih kolom yang ingin dianalisis.

````python
num_cols = ['Line Item Value']

print(df[num_cols].describe())
````

````python
import plotly.express as px

fig = px.histogram(df, x="Line Item Value", nbins=30,
                   title="Distribusi Line Item Value")
fig.show()
````

### **Analisis Bivariat & Korelasi**
Peserta bebas memilih pasangan kolom.

````python
import plotly.express as px

fig = px.scatter(df, x="Line Item Quantity", y="Line Item Value",
                 trendline="ols",
                 title="Hubungan Line Item Quantity vs Line Item Value")
fig.show()
````

````python
corr = df['Line Item Quantity'].corr(df['Line Item Value'])
print("Korelasi Quantity vs Value:", corr)
````

### **Deteksi outlier**
Peserta bebas memilih kolom numerik yang relevan.

````python
Q1 = df['Line Item Value'].quantile(0.25)
Q3 = df['Line Item Value'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Line Item Value'] < lower) | (df['Line Item Value'] > upper)]

print(f"Jumlah outlier: {len(outliers)}")
print(outliers[['Line Item Value']].head())
````

### **Kesimpulan EDA**



1. Setelah dilakukan analisis univariat, distribusi Line Item Value menunjukkan sebagian besar transaksi bernilai kecil (kurang dari Rp200.000). Data memiliki distribusi yang positively skewed.
2. Terdapat hubungan positif yang kuat antara Line Item Value dan Line Item Quantity. Semakin besar jumlah barang yang dibeli, semakin besar nilai transaksinya.
3. Pada kolom Line Item Value, terdapat 1.127 outlier.
